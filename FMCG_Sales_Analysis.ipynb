## **Exploratory  Data Analysis on FMCG Sales Dataset**

### **FMCG Dataset – Overview**

This dataset simulates daily-level transactional data for products sold in the Fast-Moving Consumer Goods (FMCG) sector between 2022 and 2024.

This is designed to support analysis of:

- Seasonality and product lifecycle
- Promotion and holiday effects
- Stockouts and inventory dynamics
- Demand forecasting and feature engineering

Each row represents one SKU on a specific day, across different sales channels and region. This project is intended for educational and portfolio-building purposes.

- Link to the dataset we are using for our practice

(https://www.kaggle.com/datasets/beatafaron/fmcg-daily-sales-data-to-2022-2024?select=FMCG_2022_2024.csv)

- The data has been manipulated after it was downloaded from Kaggle. This is because the data didn't have enough information to support our analysis. In order to be able to have a clean and consistent analysis, we added some few informations to the dataset. This eventually will make our analysis an hypothetical analysis.


Four our columns, each record includes;

- date: Date of the sales record
- sku: Stock Keeping Unit (unique product ID)
- brand: Brand name of the product
- segment: Product segment (e.g., premium, standard, economy)
- category: Product category (e.g., Beverages, Snacks)
- channel: Sales channel (e.g., Retail, E-commerce)
- region: Region where the sale occurred
- pack_type: Packaging type
- price_unit: Price per single unit
- promotion_flag: Whether the sale occurred under promotion (1 = Yes, 0 = No)
- delivery_days: Number of days taken to deliver
- stock_available: Units available in stock
- delivered_qty: Number of units delivered
- units_sold: Total units actually sold

The goal is to understand the company’s performance and uncover insights that can help improve sales strategies and stock management.

#### **Importing Libraries and Loading the Dataset**

In this section I will be importing the necessary libraries that will be needed for my data analysis and manipulation. Also I will be loading my dataset for the analysis.

```python
# importing necessary libraries

import pandas as pd
import numpy as np

# Load Dataset

df = pd.read_csv("FMCG_2022_2024.csv")

df.head()
```

```python
# to preview the last 5 rows of the dataset, I use;

df.tail()
```

#### **Data Cleaning**

Before analysis, I need to ensure that the data is properly cleaned and reliable for my analysis.

```python
# Preview data

df.info()
```

```python
# Converting multiple columns to their proper datatypes

df["date"] = df["date"].astype("datetime64[ns]")
df["sku"] = df["sku"].astype("string")
df["brand"] = df["brand"].astype("string")
df["segment"] = df["segment"].astype("string")
df["category"] = df["category"].astype("string")
df["channel"] = df["channel"].astype("string")
df["region"] = df["region"].astype("string")
df["pack_type"] = df["pack_type"].astype("string")


# Rechecking to see if my data has been properly formatted

df.info()
```

```python
# Checking for missing values

df.isnull().sum()
```

```python
# Fill missing values with 0

numeric_cols = df.select_dtypes(include=["float"]).columns

df[numeric_cols] = df[numeric_cols].fillna(0)

# Rechecking to see if dataset with Null has been updated

df.isnull().sum()
```

```python
# Converting stock_available and delivered_qty from float to integer

df["stock_available"] = df["stock_available"].astype("int64")
df["delivered_qty"] = df["delivered_qty"].astype("int64")

# Rechecking to see if my data has been properly formatted

df.info()
```

```python
#Checking for duplicates
df.duplicated().sum()
```

```python
# Drop duplicates

df.drop_duplicates(inplace=True)

df.info()
```

```python
# renaming the segment, category, channel and the pack_type colums

df.rename(columns={
    "segment": "product_segment",
    "category": "product_category",
    "channel": "product_channel",
    "pack_type": "packaging_type"
}, inplace=True)

#To preview if the columns have been properly renamed

df.info()
```

### **Data Analysis and Visualization**

#### **Data Analysis**

This is the process where we inspect, transform and model the data to discover useful information, derive conclusion, and support decision making. This stage is when we turn our cleaned data into actionable insights.

#### **Data Visualization**

This is where we perform graphical representation of our data. It is a key step in data analysis that uses visual elements like charts, graphs and maps in order to see and understand trends, outliers and patterns in our datsets.

```python
# importing libraries for our visualizations

import matplotlib.pyplot as plt
import seaborn as sns

# View dataset summary statistics 

df.describe()
```

```python
# Calculating the total sales for each record

df["total_sales"] = df["units_sold"] * df["price_unit"]

df.head()
```

```python
# 1. Checking the region with highest sales

region_sales = df.groupby("region")["total_sales"].sum().sort_values(ascending= False)
print(region_sales)
```

```python
# Matplotlib Visuals: Showing the region with highest sales

plt.figure(figsize=(4,6))  #plot size
plt.pie(region_sales, labels=region_sales.index, autopct='%1.2f%%', startangle=90)
plt.title("Total Sales by Region")
plt.show()
```

##### **Insight:**  
Sales across the three regions — PL-North, PL-South, and PL-Central — are almost identical. Each region contributes roughly 33% of total sales.

##### **Meaning:**
- The company has balanced market penetration across all regions.
- No region is significantly outperforming or underperforming.
- Growth opportunities may lie in regional strategy rather than redistribution.

```python
# 2. Check top product_categories by total sales

top_cat = df.groupby("product_category")["total_sales"].sum().sort_values(ascending=False)
print(top_cat)
```

```python
# Matplotlib Visuals: Total sales by product categories 

plt.figure(figsize=(8,4))
plt.bar(top_cat.index, top_cat.values, color = "blue")
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Using Seaborn Visuals with the same analysis: Totals by Product Category

plt.figure(figsize=(8,4))
sns.barplot(x=top_cat.index, y=top_cat.values, color = "red")
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.show()
```

##### **Insight:**
There is a dominant product category, with others performing at moderate or low levels.
Here’s the order from highest to lowest:

1. Yogurt – Significantly the highest-selling category.
2. Milk – Strong but far below Yogurt.
3. ReadyMeal – Moderate.
3. SnackBar – Moderate but slightly lower than ReadyMeal.
4. Juice – Very low sales compared to all others.

##### **Meaning:**
- Yogurt is the major revenue driver and should be prioritized in production, distribution, and marketing.
- Juice may require investigation — low demand, poor availability, or pricing issues.
- The mid-performing categories (Milk, ReadyMeal, SnackBar) offer opportunities for targeted promotions to boost performance.

```python
# 3. best brands across all sales

brands_sales = df.groupby("brand")["total_sales"].sum().sort_values(ascending=False)
brands_sales
```

```python
# Matplotlib Visuals: best brands across all sales

plt.figure(figsize=(20,10))
plt.bar(brands_sales.index, brands_sales.values, color = "purple")
plt.title("Best Performing Brands  across all Sales")
plt.xlabel("Brands")
plt.ylabel("Total Sales")
plt.show()
```

##### **Insight:**
Sales are heavily concentrated in a few top brands, with **SnBrand2 and YoBrand4** clearly dominating, while many other brands show significantly weaker performance.

##### **Meaning:**

- SnBrand2 is the strongest performer with the highest total sales.
- YoBrand4 also performs strongly and ranks second.
- YoBrand products collectively show strong market acceptance.
- Mid-tier brands (MiBrand3, YoBrand1, ReBrand4) contribute moderately.
- Several brands generate low sales, indicating weak demand or poor market penetration.
- Performance varies widely within brand families (e.g., SnBrand2 strong vs SnBrand3 weak).
- Market is top-heavy — a few brands account for most of the revenue.

```python
# 4. Matplotlib Visuals: Relationship between Unit Price and Units Sold

plt.figure(figsize=(15,8))
plt.scatter(df["price_unit"], df["units_sold"], color ="green", alpha=0.6)
plt.title("Price vs Units Sold")
plt.xlabel("Unit Price")
plt.ylabel("Unit Sold")
plt.show()
```

```python
# Using Seaborn Visuals with the same analysis: Relationship between Unit Price and Units Sold

plt.figure(figsize=(15,8))
sns.scatterplot(x= "price_unit", y="units_sold", data = df, color="orange", alpha=0.6)
plt.title("Price vs Units Sold")
plt.xlabel("Unit Price")
plt.ylabel("Units Sold")
plt.show()
```

##### **Insight:**
There is no meaningful correlation between Unit Price and Units Sold. Sales volume remains highly scattered across all price levels, showing that price has little to no impact on the number of units purchased.

##### **Meaning:**
- Units sold remain randomly distributed regardless of price level.
- Customers continue buying similar quantities even as prices increase or decrease.
- Demand is price-inelastic — buyers are not sensitive to price changes.
- The dense band between 0–60 units indicates most sales fall within that range, independent of price.
- Higher prices do not reduce sales, and lower prices do not increase sales.
- Other factors like product category, brand reputation, promotions, retailer influence, etc., likely drive sales more than price.

```python
# 5. Yearly Sales Trend

yearly_sales = df.groupby(df["date"].dt.to_period("Y"))["total_sales"].sum().sort_index()
yearly_sales
```

```python
# Matplotlib Visuals: Yearly Sales Trend

plt.figure(figsize=(8,4))
plt.plot(yearly_sales.index.astype(str), yearly_sales.values, marker="o")
plt.title("Yearly Sales Trend ")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Using Seaborn Visuals with the same analysis: Yearly Sales Trend

plt.figure(figsize=(8,4))
sns.lineplot(x=yearly_sales.index.astype(str), y=yearly_sales.values, marker='o')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

##### **Insight:**
Total sales were lowest in 2022, then increased sharply in 2023, and remained steady at a high level in 2024.

##### **Meaning:**
- 2022 recorded the weakest performance in the 3-year period.
- Sales improved significantly in 2023.
- Sales remained consistently strong from 2023 to 2024 even though 2023 recorded the year with the highest sales.
- The upward shift indicates recovery and growing demand after 2022.
- The stable high performance in 2023–2024 suggests stronger market positioning or better operational efficiency.

```python
# 6. Region with the Highest Sales After Promotion
# Filter only promotional sales

promo_sales =df[df["promotion_flag"] == 1]

# Group by region and sum revenue
region_sales = promo_sales.groupby('region')['total_sales'].sum().sort_values(ascending=False)
region_sales
```

```python
# Matplotlib Visuals: Region with the Highest Sales After Promotion

plt.figure(figsize=(4,6))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.2f%%', startangle=90)
plt.title("Total Sales by Region (After Promotion)")
plt.show()
```

##### **Insight:**
After the promotional campaign, PL-North achieved the highest sales share (33.50%), slightly outperforming PL-South (33.48%) and PL-Central (33.02%). The differences are very small, indicating that the promotion impacted all regions almost equally, with PL-North having a marginal lead.

##### **Meaning:**
- All three regions responded strongly to the promotion, with sales shares clustered closely around 33%.
- PL-North’s slight lead suggests marginally better customer engagement or market activity during the promo.
- PL-South performed almost identically, showing a nearly equal promotional effect in that region.
- PL-Central, though slightly lower, still achieved strong results, indicating that promotion effectiveness was consistent across regions.
- The minimal differences imply that the promotional strategy was evenly effective nationwide, not dependent on geography.
- This balanced response suggests that the campaign can be scaled or repeated across all regions with similar expected outcomes.

```python
# 7. Region with the Most Stock on Ground

region_stock = df.groupby("region")["stock_available"].sum().sort_values(ascending=False)
region_stock
```

```python
# Matplotlib Visuals: Total Stock Available by Region

plt.figure(figsize=(4,6))
plt.pie(region_stock, labels=region_stock.index, autopct='%1.2f%%', startangle=90)
plt.title("Total Stock Available by Region")
plt.show()
```

##### **Insight:**
Stock availability is almost evenly distributed across all regions, with PL-North holding the highest share (33.43%), followed closely by PL-South (33.32%) and PL-Central (33.25%). The difference across regions is extremely small, showing a balanced supply strategy.

##### **Meaning:**
- Stock levels are uniformly allocated across the three regions.
- PL-North has the highest stock share but only by a small margin, indicating no region is significantly prioritized over others.
- PL-South and PL-Central also have nearly equal stock, showing fair and consistent distribution.
- The tight range (33.25%–33.43%) suggests strong inventory planning and balanced supply chain operations.
- This distribution reduces the risk of regional stockouts and supports consistent product availability for customers nationwide.
- A balanced stock allocation also indicates that demand across regions is relatively similar, or the company is ensuring equal readiness for all markets.

```python
# 8. Sales Channel with the Highest Revenue

channel_revenue = df.groupby("product_channel")["total_sales"].sum().sort_values(ascending=False)
channel_revenue
```

```python
# Matplotlib Visuals: Sales Channel with the Highest Revenue

plt.figure(figsize=(4,6))
plt.pie(channel_revenue, labels=channel_revenue.index, autopct='%1.2f%%', startangle=90)
plt.title("Revenue by Sales Channel")
plt.show()
```

##### **Insight:**
Retail generated the highest revenue (33.36%), followed very closely by E-commerce (33.35%) and Discount channels (33.29%). The revenue distribution is almost equal, showing that all three channels contribute nearly the same level of sales.

##### **Meaning:**
- Retail leads by a very narrow margin, indicating strong in-store customer activity.
- E-commerce is almost tied with retail, showing that online sales are just as important and well-established.
- Discount channels also perform strongly, contributing nearly one-third of total revenue.
- The extremely small differences across channels mean the business is not dependent on a single sales channel.
- Revenue strength across all channels suggests a well-balanced omnichannel strategy.
- This balance helps reduce risk — if one channel slows down, others can maintain overall revenue.
- The company is successfully reaching customers both online and offline, indicating healthy market coverage.
